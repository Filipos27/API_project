```python
#fuction for gettin videos from channels
def get_channel_videos(channel_id):
    response=youtube.channels().list(id=channel_id,
                                    part="contentDetails").execute()
    playlist_id=response["items"][0]["contentDetails"]["relatedPlaylists"]["uploads"]
    videos=[]
    next_page_token=None
    
    while 1:
        response=youtube.playlistItems().list(playlistId=playlist_id,
                                             part="snippet",
                                             maxResults=50,
                                             pageToken=next_page_token).execute()
        videos+=response["items"]
        next_page_toen=response.get("nextPageToken")
        
        if next_page_token is None:
            break
            
    return videos
```

```python
videos=get_channel_videos("UCvgfXK4nTYKudb0rFR6noLA")
```

```python
#Function for gettin video stats 
def video_stats(video_ids):
    stats=[]
    for i in range(0,len(video_ids),50):
        response=youtube.videos().list(id=",".join(video_ids[i:i+50]),
                                      part="statistics").execute()
        stats += response["items"]
        return stats
```

```python
video_ids=list(map(lambda x:x["snippet"]["resourceId"]["videoId"],videos))
```

```python
stats=video_stats(video_ids)
```

```python
len(stats)
```

```python
#Most disliked videos
most_disliked=sorted(stats, key=lambda x:int(x["statistics"]["dislikeCount"]),reverse=True)
```

## API youtube all in one

```python
#fuction for getting in certain channel
def get_channel(channel_name):
    response=youtube.search().list(q=channel_name, type='channel', part='id,snippet').execute()['items'][0]
    return response

#function for gettin video of certain channels
def get_videos(channel_id, part='id,snippet', limit=10):
    res = youtube.channels().list(id=channel_id, 
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id, 
                                           part=part, 
                                           maxResults=min(limit, 50),
                                           pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None or len(videos) >= limit:
            break

    return videos
#setting time (timedelta is one hour because location of this channel is in UK)
def parse_publish_timestamp(video):
    return (datetime.strptime(video['snippet']['publishedAt'], "%Y-%m-%dT%H:%M:%SZ")
            + timedelta(hours=1, minutes=0))
```

```python
# Picking channel
channel=get_channel("Boxing Legends TV")
channel
```

```python
# Gettin channel id
channel_id=get_channel("Boxing Legends TV")["id"]["channelId"]
channel_id
```

```python
#gettin videos of channel
videos=get_videos(channel_id,limit=50)
```

```python
#setting timestamps for each videos
publish_timestamps = [parse_publish_timestamp(video) for video in videos]
publish_timestamps
```

```python
#Getting exact time of the date for each timestamp
publish_times=[t.hour+t.minute/60 for t in publish_timestamps]
publish_times
```
